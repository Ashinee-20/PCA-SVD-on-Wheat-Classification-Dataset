### PCA Technique

```python
mat = data.drop(['category'], axis=1)
val = data.values

mean = np.mean(val, axis= 0)
mean_data = val - mean

cov = np.cov(mean_data.T)
cov = np.round(cov, 2)

eig_val, eig_vec = np.linalg.eig(cov)

indices = np.arange(0,len(eig_val), 1)
indices = ([x for _,x in sorted(zip(eig_val, indices))])[::-1]
eig_val = eig_val[indices]
eig_vec = eig_vec[:,indices]

sum_eig_val = np.sum(eig_val)
explained_variance = eig_val/ sum_eig_val
#print(explained_variance)
cumulative_variance = np.cumsum(explained_variance)
print("cumulative variance: ", cumulative_variance)
pca_data = np.dot(mean_data, eig_vec)
```

```python
num = np.argmax(cumulative_variance > 0.99)
num
```

```python
pca_df = pd.DataFrame(pca_data, columns=[f'PC_{i+1}' for i in range(len(data.columns))])
cols = pca_df.columns
pca = pd.DataFrame()
for i in range(num):
    pca[cols[i]] = pca_df[cols[i]]
#pca['label'] = data['category']
pca
```

## SVD Implementation

```python
def custom_svd(dataframe, num_components=None):

    features = dataframe.drop(columns=['category'])
    labels = dataframe['category']

    X = features.values
    cov_matrix = np.dot(X.T, X)
    eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

    sorted_indices = np.argsort(eigenvalues)[::-1]
    eigenvalues = eigenvalues[sorted_indices]
    eigenvectors = eigenvectors[:, sorted_indices]

    sigma = np.sqrt(eigenvalues)
    U = X.dot(eigenvectors)[:, :len(sigma)]
    VT = eigenvectors.T

    if num_components is None:
        num_components = min(features.shape[0], features.shape[1])

    reduced_features = np.dot(U[:, :num_components], np.diag(sigma[:num_components]))

    reduced_dataframe = pd.DataFrame(reduced_features, columns=[f'Component_{i+1}' for i in range(num_components)])
    reduced_dataframe['category'] = labels

    return reduced_dataframe
```

```python
reduced_dataframe_svd = custom_svd(data, num_components=2)
reduced_dataframe_svd
```

```python
from sklearn.manifold import TSNE

tsne = TSNE(n_components=2, verbose=1, perplexity=40, n_iter=300)
tsne_results = tsne.fit_transform(reduced_dataframe_svd)

df = pd.DataFrame()
df['tsne-2d-one'] = tsne_results[:,0]
df['tsne-2d-two'] = tsne_results[:,1]
df['label'] = data['category']

plt.figure(figsize=(12,8))
sns.scatterplot(
    x="tsne-2d-one", y="tsne-2d-two",
    hue="label",
    data=df,
    palette=sns.color_palette("Set1"),
    legend="full",
    alpha=0.3
)
plt.title("T-SNE Vis for 3 dimention")
```
